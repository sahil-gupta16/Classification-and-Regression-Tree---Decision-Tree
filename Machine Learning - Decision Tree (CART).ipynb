# Decision Tree (CART) :
## Classification Tree :

```python
#Getting the required libraries
import numpy as np
import pandas as pd

#for train-test split
from sklearn.model_selection import train_test_split

#for fitting classification tree
from sklearn.tree import DecisionTreeClassifier

#for fitting regression tree
from sklearn.tree import DecisionTreeRegressor

#to create the confusion matrix
from sklearn.metrics import confusion_matrix

from sklearn import tree
```

```python
# Set the working directory 
import os
os.chdir(r"C:\Users\Lenovo\Documents\Codes")
#Read the iris.csv data
iris = pd.read_csv("iris.csv")
iris
```

```python
## Separating the target and the predictors
```

```python
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y= iris['species']
```

```python
## Train - test split(stratified sampling of y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, stratify = y, random_state = 100)
```

```python
#validation data(optional)
#X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.3, stratify = y_train, random_state = 100)
```

```python
#X_train, X_test, y_train, y_test
```

```python
# Decision tree classifier with criterion gini index
```

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                                 max_depth = 3, min_samples_leaf = 5)

clf_gini.fit(X_train, y_train)
```

```python
#Classifying the new observations
```

```python
#Making predictions
y_pred = clf_gini.predict(X_test)
y_pred
```

```python
# Confusion Matrix
```

```python
c = confusion_matrix(y_test, y_pred)
c
```

```python
clf_gini.score(X_test, y_test)
```
